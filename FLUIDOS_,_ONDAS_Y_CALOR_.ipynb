VELOCIDAD DE SONIDO EN DIFERENTES LIQUIDOS

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para diferentes medios
medios = ['Aire', 'Agua', 'Acero']
velocidades = [343, 1482, 5960]  # en m/s

# Gráfico de barras para velocidad del sonido en diferentes medios
plt.figure(figsize=(10, 6))
plt.bar(medios, velocidades, color=['skyblue', 'lightgreen', 'lightcoral'])
plt.xlabel('Medio')
plt.ylabel('Velocidad del Sonido (m/s)')
plt.title('Velocidad del Sonido en Diferentes Medios')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Datos para variación con la temperatura en el aire
temperaturas = np.linspace(-20, 40, 100)  # Temperaturas de -20°C a 40°C
velocidad_aire = 331.3 + 0.6 * temperaturas  # Aproximación lineal

# Gráfico de velocidad del sonido en el aire vs. temperatura
plt.figure(figsize=(10, 6))
plt.plot(temperaturas, velocidad_aire, color='purple')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Velocidad del Sonido en el Aire (m/s)')
plt.title('Velocidad del Sonido en el Aire vs. Temperatura')
plt.grid(True)
plt.show()
```
